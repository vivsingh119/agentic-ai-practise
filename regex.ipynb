````python
import regex as re;
numstr="""dsjfjfd323&dsfsfbfj32344"""
re.findall("[0-9]",numstr)
````

````python
re.findall("[^0-9]",numstr)
````

````python
#get combination of 2 number
re.findall("[0-9]{2}",numstr)
````

````python
#find the number which is greater than 200
str1="""dajkjasWnfsdf221012331dkm$#@kdf$&dfns3223nsd32120 njdfn31-022321249"""
re.findall("[2-9][1-9][1-9]",str1)
````

````python
str2="""djdjs02323323543849sdjjksfd3224395969568945694596nmcnvn54045034503490699wqosdmG%6783934848483932jghfdj786493483278dkjdsa765"""
re.findall("[6-9][0-9]{9}",str2)
````

````python
# find max 4 digit combination. i.e; it will include 1 ,2,3,4 digit combination and not more than 4 digit combination
re.findall('[0-9]{1,4}',str2)
````

````python
# find max 6 digit and min 3 digit combination. i.e; it will include 3,4,5,6 digit combination
# and not more than 6 digit  and not less than 3 digit combination
re.findall('[1-9][0-9]{2,5}',str2)
````

````python
# find the combination of digit which is minimum 3 digit and max will be n numbers
re.findall("[0-9]{3,}",str2)
````

````python
# minimum 1 max n numbers
re.findall("[0-9]+",str2)
````

````python
email=input("Enter a vaild Email: ")
emailRes=re.fullmatch("[a-zA-Z0-9_.]+@[a-zA-Z0-9]+[.][a-z]{2,}",email)
print("Valid Email ID" if emailRes else "Invalid Email Id")
````
